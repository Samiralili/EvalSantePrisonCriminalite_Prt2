```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
pol = pd.read_csv('crim_plce_1_Data.csv', sep=",")
pol.drop(['UNIT', 'Flag and Footnotes'], axis=1, inplace=True)
pol.dropna(axis=0,inplace=True)
pol.Value = pol.Value.astype(int)
pol=pol.sort_values(by='Value',ascending=False)
pol
```

```python
inf = pd.read_csv("crim_gen_1_Data.csv", sep=",")
inf.drop(columns="Flag and Footnotes", inplace = True)
inf.dropna(axis=0,inplace=True)
inf
```

```python
inf2 = inf[inf['GEO'] == 'Turquie']
inf2 = inf2.drop(inf2[inf2['ICCS'] == 'Total'].index)
inf3 = inf2.groupby(['TIME','GEO'])['Value'].sum()
inf4 = pd.DataFrame(inf3)
inf4.reset_index(inplace=True)
```

```python
inf4
```

```python
plt.figure(figsize=(12,9))
ax = plt.plot(inf4.TIME,inf4.Value, color='green')
ax1 = plt.plot(pol2.TIME,pol2.Value, color='red')
plt.ylabel("Valeurs",color='black')
plt.xlabel("Années", color='black')
plt.xticks(color='grey')
plt.yticks(color='grey')
plt.title("Lien entre nombre de policiers et nombre d’infractions", color='black')
plt.legend(('infractions','nombre de policiers'),loc='upper right')
plt.show()
```
