```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
pri = pd.read_csv('crim_pris_pop_1_Data.csv', sep=",")
pri
```

```python
pri.reset_index(inplace=True)
pri.rename(columns={"index":"TIME","TIME":"GEO","GEO":"UNIT","UNIT":"Value","Value":"Drop"},inplace=True)
pri.drop(columns=["Drop"],inplace = True)
pri.dropna(axis=0,inplace=True)
pri
```

```python
prifr = pri[pri["GEO"] == "France"]
prifr
```

```python
plt.figure(figsize=(12,9))
ax = plt.plot(prifr.TIME,prifr.Value, color='Blue')
plt.ylabel("Nombre de prisonniers",color='Black')
plt.xlabel("Années", color='Black')
plt.xticks(color='grey')
plt.yticks(color='grey')
plt.title("Evolution de la population Carcérale en france", color='Black')
plt.axis([1998,2007, None, None])
plt.show()
```
