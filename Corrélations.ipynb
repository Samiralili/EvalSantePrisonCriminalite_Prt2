## I. Nombre de policiers et criminalité

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
inf = pd.read_csv("crim_gen_1_Data.csv", sep=",")
inf.drop(columns="Flag and Footnotes", inplace = True)
inf.dropna(axis=0,inplace=True)
inf
```

```python
pol = pd.read_csv('crim_plce_1_Data.csv', sep=",")
pol.drop(['UNIT', 'Flag and Footnotes'], axis=1, inplace=True)
pol.dropna(axis=0,inplace=True)
pol.Value = pol.Value.astype(int)
#pol=pol.sort_values(by='Value',ascending=False)
poli=pd.DataFrame(pol.groupby(["GEO","TIME"])["Value"].sum())
poli
```

```python
inf = inf[inf["ICCS"] != "Total"]
infcor = inf.drop(columns="ICCS")
infcore = pd.DataFrame(infcor.groupby(["GEO","TIME"])["Value"].sum())
infcore
```

```python
core = poli.merge(infcore, left_on='GEO', right_on='GEO', how='outer')
core.dropna(axis=0, inplace=True)
core.rename(columns={"Value_x":"Policiers","Value_y":"Crimes"},inplace=True)
core
```

```python
plot = sb.regplot(core.Policiers,core.Crimes)
```

## II. Criminalité, nombre de policiers et nombre de détenus

```python
pri = pd.read_csv('crim_pris_pop_1_Data.csv', sep=",")
pri.reset_index(inplace=True)
pri.rename(columns={"index":"TIME","TIME":"GEO","GEO":"UNIT","UNIT":"Value","Value":"Drop"},inplace=True)
pri.drop(columns=["Drop"],inplace = True)
#pri.drop(columns=["UNIT"],inplace=True)
pri.dropna(axis=0,inplace=True)
pris=pd.DataFrame(pri.groupby(["GEO","TIME"])["Value"].sum())
pris
```

```python
core = core.merge(pris, left_on='GEO', right_on='GEO', how='outer')
core.dropna(axis=0,inplace=True)
core.rename(columns={"Value":"Détenus"},inplace=True)
core
```

```python
plot = sb.regplot(core.Détenus,core.Crimes)
```

```python
plot = sb.regplot(core.Policiers,core.Détenus)
```

## III. Dépenses de santé et criminalité

```python
sante = pd.read_csv('spr_exp_sum_1_Data.csv')
sante.dropna(axis=0, inplace=True)
sante = sante[sante["SPDEPS"] == "Maladie/Soins de santé"]
sante = sante[sante["UNIT"] == "Millions d'euros"]
sante.drop(columns=['Flag and Footnotes', 'SPDEPS'], inplace=True)
sante.drop(columns=['UNIT'], inplace=True)

sante["Value"] = sante["Value"].str.replace(",",".")
sante["Value"] = sante["Value"].astype(float)
sante=pd.DataFrame(sante.groupby(["GEO","TIME"])["Value"].sum())
sante
```

```python
core =sante.merge(infcore, left_on=['GEO'], right_on=['GEO'], how='outer')
core.dropna(axis=0, inplace=True)
core.rename(columns={"Value_x":"Dépenses_santé","Value_y":"Crimes"},inplace=True)
core
```

```python
plot = sb.regplot(core.Dépenses_santé,core.Crimes)
```

## IV. Criminalité et densité

```python
pop = pd.read_csv('demo_pjan_1_Data.csv', sep=",")
sup = pd.read_csv('EuropeSuperficie.csv', sep=",")
demo = sup.merge(pop, left_on='États_membres', right_on='GEO', how='outer')
demo.drop(['AGE', 'SEX', 'UNIT', 'Flag and Footnotes'], axis=1, inplace=True)
demo.dropna(axis=0,inplace=True)
demo = demo[demo.Value != ":"]
demo["Value"] = demo.Value.dropna(axis=0)
demo["Value"] = demo["Value"].astype(int)
demo['Densité']=demo['Value']/demo['Superficie_(km2)']
demo.drop(['États_membres', 'Superficie_(km2)', 'Value'], axis=1, inplace=True)
demo
```

```python
Densité=pd.DataFrame(demo.groupby(["GEO","TIME"]).sum())
Densité
```

```python
core =Densité.merge(infcore, left_on=['GEO'], right_on=['GEO'], how='outer')
core.dropna(axis=0, inplace=True)
core
```

```python
core.rename(columns={"Value":"Crimes"},inplace=True)
core
```

```python
plot = sb.regplot(core.Densité,core.Crimes)
```
